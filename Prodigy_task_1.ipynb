```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a sample dataset (you can replace this with your own dataset)
data = pd.DataFrame({'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female', 'Female']})

# Create a bar chart to visualize the distribution of the 'Gender' variable
sns.countplot(x='Gender', data=data)
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
